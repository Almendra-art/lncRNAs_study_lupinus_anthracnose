```r
#This step is to generate a merged big file of your desire quantil-filtered network, including pairs genes connections attributes: weights and treatment-related
#input files were obtained from run12
#In our study input files are related to two genotypes (r and s) and different hours post infection (hpi)

merge_big_r <- read.delim("your_path/bignet_r_099.txt", sep ='\t', header=TRUE)
merge_big_r24hpi <- read.delim("your_path/bignet_r24hpi_099.txt", sep ='\t', header=TRUE)
merge_big_r60hpi <- read.delim("your_path/bignet_r60hpi_099.txt", sep ='\t', header=TRUE)
merge_big_r84hpi <- read.delim("your_path/bignet_r84hpi_099.txt", sep ='\t', header=TRUE)

merge_big_s <- read.delim("your_path/099/bignet_s_099.txt", sep ='\t', header=TRUE)
merge_big_s24hpi <- read.delim("your_path/099/bignet_s24hpi_099.txt", sep ='\t', header=TRUE)
merge_big_s60hpi <- read.delim("your_path/099/bignet_s60hpi_099.txt", sep ='\t', header=TRUE)
merge_big_s84hpi <- read.delim("your_path/099/bignet_s84hpi_099.txt", sep ='\t', header=TRUE)

```

```r
#clean columns
dfs <- ls(pattern = "^merge")
for (d in dfs) {
  tmp <- get(d)
  tmp <- tmp[, !(names(tmp) %in% c("fromAltName", "toAltName", "direction"))]
  assign(d, tmp)
}
```

```r
#add pairs information to pairs associated with bignet fi
#solo tablas con pairs identificados en la bignet
merge_big_r$pair <- "r" 
merge_big_r24hpi$pair <- "r24hpi"
#merge_big_r60hpi$pair <- "r60hpi"
merge_big_r84hpi$pair <- "r84hpi"

merge_big_s$pair <- "s" 
merge_big_s24hpi$pair <- "s24hpi" 
#merge_big_s60hpi$pair <- "s60hpi"
#merge_big_s84hpi$pair <- "s84hpi"
```

```r
#concatenar
merge_pairs <- rbind(merge_big_r, merge_big_r24hpi,merge_big_r84hpi,
                          merge_big_s, merge_big_s24hpi)
```
